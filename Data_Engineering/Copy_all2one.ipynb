```python
import pandas as pd
import numpy as np
import gc
import os, shutil
import sys
import matplotlib.pyplot as plt
import open3d as o3d
from tqdm import tqdm_notebook
```

```python
BASE_PATH = "/home/sabyasachi/Projects/ati/data/data/datasets/Carla/few_dynamic_runs/110k"
INPUT_FOLDER = "static"
INPUT_PATH = os.path.join(BASE_PATH, INPUT_FOLDER)
NEW_OUT_FOLDER = INPUT_FOLDER +"_all"

PCD_FOLDER = "_out"
PCD_TYPE = ".ply"

NEW_OUT_PATH = os.path.join(BASE_PATH, NEW_OUT_FOLDER)
if not os.path.exists(NEW_OUT_PATH):
    os.makedirs(NEW_OUT_PATH)
else:
    shutil.rmtree(NEW_OUT_PATH)
    os.makedirs(NEW_OUT_PATH)
```

```python
file_idx = 0
for subfolder in tqdm_notebook(os.listdir(INPUT_PATH)):
    pcd_path = os.path.join(INPUT_PATH, subfolder, PCD_FOLDER)
    for pcd_file in os.listdir(pcd_path):
        src_path = os.path.join(pcd_path, pcd_file)
        dst_fname = str(file_idx) + PCD_TYPE
        dst_path = os.path.join(NEW_OUT_PATH, dst_fname)
        shutil.copy(src_path,dst_path)
        file_idx += 1
```

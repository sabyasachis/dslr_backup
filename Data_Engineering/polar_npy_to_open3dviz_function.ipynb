```python
# Helper functions
def draw_pcd(pcd, where='mat_3d'):
    if where is 'opn_nb':
        visualizer = o3d.JVisualizer()
        visualizer.add_geometry(pcd)
        visualizer.show()
    elif where is 'opn_view':
        o3d.visualization.draw_geometries([pcd], width=1280, height=800)
    elif where is 'mat_3d':
        plt.figure()
        pts = np.asarray(pcd.points)
        plt.scatter(pts[:,0], pts[:,1], pts[:,2])
        plt.grid()
        plt.show()
    elif where is 'mat_2d':
        plt.figure()
        pts = np.asarray(pcd.points)
        plt.scatter(pts[:,0], pts[:,1])
        plt.grid()
        plt.show()
        
def from_polar_np(velo):
    angles = np.linspace(0, np.pi * 2, velo.shape[-1])
    dist, z = velo[:, 0], velo[:, 1]
    x = np.cos(angles) * dist
    y = np.sin(angles) * dist
    out = np.stack([x,y,z], axis=1)
    return out.astype('float32')

def get_pcd(idx, pcd_arr):
    this_arr = pcd_arr[idx].reshape(3,-1).T
    this_pcd = o3d.geometry.PointCloud()
    this_pcd.points = o3d.utility.Vector3dVector(this_arr)
    return this_pcd

# Save polar npy to ply
def npy2ply(polar_npy_arr, output_path):
    npy_3d_arr = from_polar_np(polar_npy_arr)  # convert polar to non-polar
    npy_3d_arr_orig = npy_3d_arr * 120   # rescale from -1,1 back to -120,120
    for npy_idx in tqdm_notebook(range(polar_npy_arr.shape[0])):
        the_pcd = get_pcd(npy_idx, npy_3d_arr_orig) # get pcd from arr
        pcd_fname = str(npy_idx+1) + ".ply"
        pcd_path = os.path.join(output_path, pcd_fname)
        o3d.io.write_point_cloud(pcd_path, the_pcd)
        

# Polar npy arr to open3d vizualization in jupyter notebook
def npy2viz(polar_npy_arr, npy_idx, mask_arr=None, scale=False):
    npy_3d_arr = from_polar_np(polar_npy_arr)
    if scale:
        npy_3d_arr_orig = npy_3d_arr * 120
    else:
        npy_3d_arr_orig = npy_3d_arr
    the_pcd = get_pcd(npy_idx, npy_3d_arr_orig)
    draw_pcd(the_pcd, where='opn_nb')
    

    
```
